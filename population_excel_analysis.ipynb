```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\Bhavana B V\OneDrive\Documents\SKILLS\data.xls"

data = pd.read_excel(
    file_path,
    skiprows=3,
    engine="xlrd"
)

print("Columns in the dataset:", data.columns)
```

```python
year = "2024"

values = pd.to_numeric(data[year], errors='coerce').dropna()

print(f"First 10 values for {year}:", values.head(10))
```

```python
plt.figure(figsize=(10,6))
plt.hist(values, bins=15, color='skyblue', edgecolor='black')
plt.xlabel(f"Population Value ({year})")
plt.ylabel("Frequency")
plt.title(f"Distribution of Population Values in {year}")
plt.grid(axis='y', alpha=0.75)
plt.show()
```
